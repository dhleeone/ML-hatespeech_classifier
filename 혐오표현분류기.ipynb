```python
from konlpy.tag import Mecab 
Mecab = Mecab(dicpath=r"C:\mecab\mecab-ko-dic")
from konlpy.tag import Okt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
```

```python
gender = pd.read_excel("C:/ai_data/9m/advancedProject/comments/comment_gender.xlsx")
queer = pd.read_excel("C:/ai_data/9m/advancedProject/comments/comment_queer.xlsx")
birth = pd.read_excel("C:/ai_data/9m/advancedProject/comments/comment_지역.xlsx")
race = pd.read_excel("C:/ai_data/9m/advancedProject/comments/comment_인종.xlsx")
politics = pd.read_excel("C:/ai_data/9m/advancedProject/comments/comment-정치.xlsx")
generation = pd.read_excel("C:/ai_data/9m/advancedProject/comments/comment-generation.xlsx")
religion = pd.read_excel("C:/ai_data/9m/advancedProject/comments/comment-religion.xlsx")

list_all = [gender, queer, birth,race ,politics ,generation ,religion]
```

```python
######
tokenizer = Mecab
def mecab_tokenizer(text):
  return tokenizer.morphs(text)

######
list_X = []
list_y = []
list_tfidf = []
for i in range(0, len(list_all)):

  X = list_all[i]["contents"]
  y = list_all[i]["label"]

  tfidf = TfidfVectorizer(tokenizer=mecab_tokenizer,
                        ngram_range=(1,2), min_df=5, max_df=0.9) 
  df_tfidf = tfidf.fit_transform(X)
  list_tfidf.append(tfidf)
  list_X.append(df_tfidf)
  list_y.append(y)
```

```python
# linearSVC

def gender_classifier(mes):
  svm = LinearSVC(C=0.21)
  svm.fit(list_X[0], list_y[0])
  mes_tfidf = list_tfidf[0].transform(mes)
  return svm.predict(mes_tfidf)

def queer_classifier(mes):
  svm = LinearSVC(C=0.21)
  svm.fit(list_X[1], list_y[1])
  mes_tfidf = list_tfidf[1].transform(mes)
  return svm.predict(mes_tfidf)

def birth_classifier(mes):
  svm = LinearSVC(C=0.21)
  svm.fit(list_X[2], list_y[2])
  mes_tfidf = list_tfidf[2].transform(mes)
  return svm.predict(mes_tfidf)

def race_classifier(mes):
  svm = LinearSVC(C=0.21)
  svm.fit(list_X[3], list_y[3])
  mes_tfidf = list_tfidf[3].transform(mes)
  return svm.predict(mes_tfidf)

def politics_classifier(mes):
  svm = LinearSVC(C=0.21)
  svm.fit(list_X[4], list_y[4])
  mes_tfidf = list_tfidf[4].transform(mes)
  return svm.predict(mes_tfidf)

def generation_classifier(mes):
  svm = LinearSVC(C=0.21)
  svm.fit(list_X[5], list_y[5])
  mes_tfidf = list_tfidf[5].transform(mes)
  return svm.predict(mes_tfidf)

def religion_classifier(mes):
  svm = LinearSVC(C=0.21)
  svm.fit(list_X[6], list_y[6])
  mes_tfidf = list_tfidf[6].transform(mes)
  return svm.predict(mes_tfidf)
```

```python
def total(mes):
    
    list_filter = []

    gender_result = gender_classifier(mes)
    queer_result = queer_classifier(mes)
    birth_result = birth_classifier(mes)
    race_result = race_classifier(mes)
    politics_result = politics_classifier(mes)
    generation_result = generation_classifier(mes)
    religion_result = religion_classifier(mes)

    if gender_result == [0]:
        pass
    elif gender_result == [1]:
        list_filter.append("남성")
    elif gender_result == [2]:
        list_filter.append("여성")

    if queer_result == [0]:
        pass
    elif queer_result == [1]:
        list_filter.append("성소수자")

    if birth_result == [0]:
        pass
    elif birth_result == [1]:
        list_filter.append("수도권")
    elif birth_result == [2]:
        list_filter.append("비수도권")

    if race_result == [0]:
        pass
    elif race_result == [1]:
        list_filter.append("흑인")
    elif race_result == [2]:
        list_filter.append("백인")
    elif race_result == [3]:
        list_filter.append("동양인")

    if politics_result == [0]:
        pass
    elif politics_result == [1]:
        list_filter.append("보수")
    elif politics_result == [2]:
        list_filter.append("진보")

    if generation_result == [0]:
        pass
    elif generation_result == [1]:
        list_filter.append("젊은세대")
    elif generation_result == [2]:
        list_filter.append("기성세대")

    if religion_result == [0]:
        pass
    elif religion_result == [1]:
        list_filter.append("종교")

    if len(list_filter) > 0:
        print(", ".join(list_filter) + "에 대한 혐오표현이 담긴 댓글입니다.")
    elif len(list_filter) == 0:
        print("혐오표현이 없습니다.")
```

```python
mes = input("댓글을 입력하세요. > ")

total([mes])
```
